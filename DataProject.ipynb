```python
films = pd.read_csv("movie_metadata.csv")
```

```python
films.head()
```

```python
films.columns
```

```python
films["profit"] = (films["gross"] - films["budget"])

films.head()
```

```python
allGenres = ""
for g in films.genres:
    allGenres += "|"+g
genre_set = list(set(allGenres.split("|")))
genre_set
```

```python
for genre in genre_set:
    if genre != "":
        films[genre] = [genre in g for g in films["genres"]]
        
films.head()
```

```python
genre_set.remove('')
```

```python
genreGrossTotalList = []

for genre in genre_set:
    genreTotalGross = 0.0
    for index in range(len(films[genre])):
        if films[genre][index] and not math.isnan(films["gross"][index]):
            genreTotalGross += float(films["gross"][index])
    genreGrossTotalList.append([genre, genreTotalGross])
    
genreGrossTotalList
```

```python
genreGrossTotalListDataFrame = pd.DataFrame(genreGrossTotalList, columns=["Genre", "Total Earnings"])
genreGrossTotalListDataFrame.plot(kind="bar", x="Genre", y="Total Earnings", color=(0.8,0.2,0.2))
```

```python
genreGrossAverageList = []

for genre in genre_set:
    genreTotalGross = 0.0
    filmGenreCount = 0.0
    for index in range(len(films[genre])):
        if films[genre][index] and not math.isnan(films["gross"][index]):
            genreTotalGross += float(films["gross"][index])
            filmGenreCount += 1
    try:
        genreGrossAverageList.append([genre, genreTotalGross/filmGenreCount])
    except Exception:
        genreGrossAverageList.append([genre, 0])

genreGrossAverageList
```

```python
genreGrossAverageListDataFrame = pd.DataFrame(genreGrossAverageList, columns=["Genre", "Average Earnings"])
genreGrossAverageListDataFrame.plot(kind="bar", x="Genre", y="Average Earnings",color=(0.8,0.2,0.2)
)
```
